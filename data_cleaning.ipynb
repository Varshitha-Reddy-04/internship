```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['math score'], kde=True)
plt.title("Math Score Distribution")
plt.show()
```

Most students scored between 60–80 in math, indicating average performance.

```python
sns.boxplot(data=df[['math score','reading score','writing score']])
plt.title("Score Comparison")
plt.show()
```

Math scores show more variation compared to reading and writing scores.

```python
sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Reading and writing scores are strongly correlated, indicating similar performance patterns.

## Final Insights

- The dataset contains no missing values or duplicates, indicating clean data.
- Student performance in reading and writing is strongly correlated.
- Math scores show higher variation compared to other subjects.
- Most students fall in the average score range (60–80).
