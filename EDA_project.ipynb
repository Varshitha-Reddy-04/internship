```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
os.getcwd()
```

```python
df = pd.read_csv(r"C:\Users\VARSHITHA\StudentsPerformance.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
df.dropna(inplace=True)
```

```python
sns.histplot(df['gender'])
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

There is a strong positive correlation between math score and reading score.
This indicates that students who perform well in math also tend to perform well in reading.

```python
sns.scatterplot(x='math score', y='reading score', data=df)
plt.show()
```

The scatter plot shows a clear upward trend, confirming strong positive correlation.

```python
sns.boxplot(x='gender', y='math score', data=df)
plt.show()
```

Students who completed test preparation scored higher on average.

```python
sns.boxplot(data=numeric_df)
plt.show()
```

This exploratory data analysis revealed several meaningful patterns in student performance. A strong positive correlation was observed between reading and writing scores, indicating that students who perform well in one tend to perform well in the other. Test preparation was found to significantly improve scores across all subjects. Additionally, factors such as lunch type and gender showed noticeable influence on performance, suggesting the impact of socioeconomic and behavioral factors on academic outcomes. Overall, the analysis highlights that both academic preparation and external conditions play a crucial role in student success.
